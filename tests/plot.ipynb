```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
times_ = [0]*2
flop_ = [0]*2
size_ = [0]*2

# init
times = np.array([])
flop = np.array([])
size = np.array([])
# append results
for range_ in [(1, 151, 1), (151, 301, 5), (301, 401, 10), (401, 601, 50) , (601, 801, 50), (800, 1001, 100)]:
    file_name = f"BinetAlgorithm-n{range_[1]-1}-"
    df_np = pd.read_csv("data/" + file_name, index_col=0).to_numpy()
    times = np.concatenate((times, df_np[0]))
    flop = np.concatenate((flop, df_np[1]))
    size = np.concatenate((size, np.arange(*range_)))


# Plot
plt.title("Binet algorithm - time")
plt.plot(size, times, color='blue', label='Binet algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

# Plot
plt.title("Binet algorithm - FLOPS")
plt.plot(size, flop/times, color='blue', label='Binet algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('FLOPS')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

times_[0] = times
flop_[0] = flop
size_[0] = size
```

```python
# init
times = np.array([])
flop = np.array([])
size = np.array([])
# append results

for range_ in [(1, 151, 1), (151, 301, 5), (300, 551, 50),(600, 1001, 100)]:
    file_name = f"StrassenAlgorithm-n{range_[1]-1}-"
    df_np = pd.read_csv("data/" + file_name, index_col=0).to_numpy()
    times = np.concatenate((times, df_np[0]))
    flop = np.concatenate((flop, df_np[1]))
    size = np.concatenate((size, np.arange(*range_)))


# Plot
plt.title("Strassen algorithm - time")
plt.plot(size, times, color='red', label='Strassen algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

# Plot
plt.title("Strassen algorithm - FLOPS")
plt.plot(size, flop/times, color='red',  label='Strassen algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('FLOPS')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

times_[1] = times
flop_[1] = flop
size_[1] = size
```

```python
# Plot
plt.title("Comparison of matrix multiplication algorithms - time")
plt.plot(size_[0], times_[0], color='blue', label='Binet algorithm')
plt.plot(size_[1], times_[1], color='red', label='Strassen algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

# Plot
plt.title("Comparison of matrix multiplication algorithms - FLOPS")
plt.plot(size_[0], flop_[0]/times_[0], color='blue', label='Binet algorithm')
plt.plot(size_[1], flop_[1]/times_[1], color='red', label='Strassen algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('FLOPS')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()
```
