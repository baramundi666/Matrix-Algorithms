```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = "BinetAlgorithm-n10-"
df = pd.read_csv("data/" + file_name, index_col=0)
df_np = df.to_numpy()
times = df_np[0]
flop = df_np[1]
size = np.arange(1, len(times)+1)


# Plot
plt.title("Binet algorithm - time")
plt.plot(size, times, color='blue', label='Binet algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

# Plot
plt.title("Binet algorithm - FLOPS")
plt.plot(size, flop/times, color='blue', label='Binet algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('FLOPS')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()
```
