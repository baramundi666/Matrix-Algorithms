```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# init
times = np.array([])
flop = np.array([])
size = np.array([])
#1
n = 150
file_name = f"BinetAlgorithm-n{n}-"
df_np = pd.read_csv("data/" + file_name, index_col=0).to_numpy()
times = np.concatenate((times, df_np[0]))
flop = np.concatenate((flop, df_np[1]))
size = np.concatenate((size, np.arange(1, n+1)))
#2
n = 300
file_name = f"BinetAlgorithm-n{n}-"
df_np = pd.read_csv("data/" + file_name, index_col=0).to_numpy()
times = np.concatenate((times, df_np[0]))
flop = np.concatenate((flop, df_np[1]))
size = np.concatenate((size, np.arange(151, n+1, 5)))


# Plot
plt.title("Binet algorithm - time")
plt.plot(size, times, color='blue', label='Binet algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

# Plot
plt.title("Binet algorithm - FLOPS")
plt.plot(size, flop/times, color='blue', label='Binet algorithm')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('FLOPS')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()
```
