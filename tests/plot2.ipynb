```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
times_ = [0]*2
flop_ = [0]*2
size_ = [0]*2

# init
times = np.array([])
flop = np.array([])
size = np.array([])
# append results
for range_ in [(1, 500, 1)]:
    file_name = f"Inverse-n{range_[1]-1}-"
    df_np = pd.read_csv("data/" + file_name, index_col=0).to_numpy()
    if len(df_np[0]) == (range_[1] - range_[0]):
        times = np.concatenate((times, df_np[0]))
        flop = np.concatenate((flop, df_np[1]))
        size = np.concatenate((size, np.arange(*range_)))
    else:
        min_len = min(len(df_np[0]), range_[1] - range_[0])
        times = np.concatenate((times, df_np[0][:min_len]))
        flop = np.concatenate((flop, df_np[1][:min_len]))
        size = np.concatenate((size, np.arange(range_[0], range_[0] + min_len)))


# Plot
plt.title("Inversion - time")
plt.plot(size, times, color='blue', label='Inversion')
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

# Plot
plt.title("Inversion - FLOPs")
plt.plot(size, flop, color='blue', label='Inversion')
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('FLOPs')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

times_[0] = times
flop_[0] = flop
size_[0] = size
```
