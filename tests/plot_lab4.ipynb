```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
# init
df = pd.read_csv("data/matrix_matrix_mult.csv", header=None)
size = np.array(list(map(float, df[0].to_numpy()[1:])))
time = np.array(list(map(float, df[1].to_numpy()[1:])))

log_size = np.log(size)
log_time = np.log(time)

slope, intercept, r_value, p_value, std_err = linregress(log_size, log_time)

b = slope
a = np.exp(intercept)

N = np.linspace(min(size), max(size), 100)
y = a * (N ** b)

# Plot
plt.title("Compressed matrix multiplication - time")
plt.plot(size, time, "o", color='red', label='matrix_matrix_mult')
# plt.plot(N, y, color='blue', label='y = {:.6f} * x^({:.6f})'.format(a, b))
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()

df = pd.read_csv("data/matrix_vector_mult.csv", header=None)
size = np.array(list(map(float, df[0].to_numpy()[1:])))
time = np.array(list(map(float, df[1].to_numpy()[1:])))

log_size = np.log(size)
log_time = np.log(time)

slope, intercept, r_value, p_value, std_err = linregress(log_size, log_time)

b = slope
a = np.exp(intercept)

N = np.linspace(min(size), max(size), 100)
y = a * (N ** b)

# Plot
plt.title("Compressed matrix by vector multiplication - time")
plt.plot(size, time, "o",color='red', label='matrix_vect_mult')
# plt.plot(N, y, color='blue', label='y = {:.6f} * x^({:.6f})'.format(a, b))
# plt.xscale('log')
# plt.yscale('log')
plt.xlabel('matrix size')
plt.ylabel('running time [s]')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
           fancybox=True, shadow=True, ncol=1)
plt.grid()
plt.show()
```
